```python
# Performance of no-noise model as a function of noise

sigeff0 = np.array([94.0, 
                    75.6, 
                    63.7, 
                    55.9,
                    43.5])
sigeff0_err = np.array([0.3, 
                        1.45, 
                        1.67,
                        1.52,
                        1.33])

datred0 = np.array([33.3, 
                    44.6, 
                    51.2,
                    55.8,
                    63.8])
datred0_err = np.array([0.5, 
                        1.02, 
                        1.67, 
                        1.17,
                        1.04])
```

```python
# Performance of dedicated model for given noise value, as a function of noise
sigeff2 = np.array([94.0, 
                    93.0, 
                    92.5, 
                    92.2,
                    91.6])
sigeff2_err = np.array([0.3, 
                        0.6, 
                        0.74, 
                        0.76,
                        0.5])

datred2 = np.array([33.3, 
                    32.2, 
                    31.9, 
                    31.7,
                    30.5])
datred2_err = np.array([0.5, 
                        1.17, 
                        1.14, 
                        1,
                        0.72])
```

```python
fig, (ax0, ax1) = plt.subplots(1,2,figsize=(8,3))
plt.tight_layout()

ax0.errorbar(x=noise_values,y=sigeff0,yerr=sigeff0_err, marker='.',linestyle=':', capsize=5,
             label='Original')
ax0.errorbar(x=noise_values,y=sigeff2,yerr=sigeff2_err, marker='.',linestyle='-', capsize=5,
             label='Retrained')
# ax0.errorbar(x=noise_values,y=sigeff3,yerr=sigeff2_err, marker='.',linestyle='--', capsize=5,
             # label='Retrained + threshold')
ax0.set_ylim(0,100)
ax0.set_ylabel('Signal Efficiency [%]')

ax1.errorbar(x=noise_values,y=datred0,yerr=datred0_err, marker='.',linestyle=':', capsize=5,
             label='Original',)
ax1.errorbar(x=noise_values,y=datred2,yerr=datred2_err, marker='.',linestyle='-', capsize=5,
             label='Retrained',)
# ax1.errorbar(x=noise_values,y=datred3,yerr=datred2_err, marker='.',linestyle='--', capsize=5,
#              label='Retrained + threshold')
ax1.set_ylim(0,100)
ax1.set_ylabel('Data reduction [%]')

ax0.set_xlabel(r'$\sigma_\text{noise} \; [e^-]$')
ax1.set_xlabel(r'$\sigma_\text{noise} \; [e^-]$')

#plt.xticks([2011,2015,2022,2030])
ax1.legend(frameon=False)
```
